```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2021-07-06&to=2021-07-09&guests=2&page=1"
driver.get(url)
time.sleep(6)
```

```python
time.sleep(4)
# creating empty lists for scraping data
ran_title=[]
rank=driver.find_elements_by_xpath("//h2[@class='title title-6']")
for i in rank:
    if i.text is None :
        ran_title.append("--") 
    else:
        ran_title.append(i.text)
```

```python
len(ran_title)
```

```python
City_Distance=[]
time.sleep(4)
Distance=driver.find_elements_by_xpath("//span[@class='description']")
for i in Distance:
    if i.text is None :
        City_Distance.append("--") 
    else:
        City_Distance.append(i.text)
```

```python
len(City_Distance)
```

```python
Rating=[]
time.sleep(4)
rating=driver.find_elements_by_xpath("//div[@class='score orange big']")
for i in rating:
    if i.text is None :
        Rating.append("No Rating") 
    else:
        Rating.append(i.text)
```

```python
len(Rating)
```

```python
Total_Reviews=[]
time.sleep(4)
reviews=driver.find_elements_by_xpath("//div[@class='reviews']")
for i in reviews:
    if i.text is None :
        Total_Reviews.append("--") 
    else:
        Total_Reviews.append(i.text)
```

```python
len(Total_Reviews)
```

```python
Private_from_price=[]
time.sleep(4)
price=driver.find_elements_by_xpath("//div[@class='price title-5']")
for i in price:
    if i.text is None :
        Private_from_price.append("--") 
    else:
        Private_from_price.append(i.text.split(','))
PRIVATE_FROM_PRICE=Private_from_price[0::2]
```

```python
Droms_from_price=[]
time.sleep(4)
price=driver.find_elements_by_xpath("//div[@class='price title-5']")
for i in price:
    if i.text is None :
        Droms_from_price.append("--") 
    else:
        Droms_from_price.append(i.text.split(','))
DROMS_FROM_PRICE=Droms_from_price[1::2]
```

```python
Property_description=[]
url=driver.find_elements_by_xpath("//h2[@class='title title-6']//a")
for i in url:
    Property_description.append(i.get_attribute('href'))
Property_description
```

```python
Property_desc=[]
for i in Property_description:
    driver.get(i)
    title=driver.find_elements_by_xpath("//div[@class='description-container']")
    for i in title:
        if i.text is None:
            Property_desc.append("No_Job_Description")
        else:
            Property_desc.append(i.text.split(','))
```
