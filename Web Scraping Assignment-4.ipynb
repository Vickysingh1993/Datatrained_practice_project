Question No. 6 Scrape the details of top 100 songs on billboard.com.

Url = https://www.billboard.com/
    
You have to find the following details:
    
A) Song name

B) Artist name

C) Last week rank

D) Peak rank

E) Weeks on board

Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
driver=webdriver.Chrome('chromedriver.exe')
time.sleep(4)
driver.get("https://www.billboard.com")
time.sleep(5)
Drop_button = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/header/div/ul/li[1]/a')       
Drop_button.click()

Hot_100= driver.find_element_by_xpath('/html/body/main/div[2]/div/div[1]/a/div[2]/div[1]')       
Hot_100.click()


Song_name=[]
try:
    element=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
    for i in element:
        if i.text is None:
            Song_name.append('None')
        else:
            Song_name.append(i.text)
except NoSuchElementException:
        Song_name.append('-')
        
Artist_name=[]
try:
    element=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
    for i in element:
        if i.text is None:
            Artist_name.append('None')
        else:
            Artist_name.append(i.text)
except NoSuchElementException:
        Artist_name.append('-')
        

Last_week_rank=[]
try:
    element=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
    for i in element:
        if i.text is None:
            Last_week_rank.append('None')
        else:
            Last_week_rank.append(i.text)
except NoSuchElementException:
        Last_week_rank.append('-')
        
Peak_rank=[]
try:
    element=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
    for i in element:
        if i.text is None:
            Peak_rank.append('None')
        else:
            Peak_rank.append(i.text)
except NoSuchElementException:
        Peak_rank.append('-')
        
Weeks_on_board=[]
try:
    element=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
    for i in element:
        if i.text is None:
            Weeks_on_board.append('None')
        else:
            Weeks_on_board.append(i.text)
except NoSuchElementException:
        Weeks_on_board.append('-')
        
time.sleep(3)    
data = pd.DataFrame({'Song_name':Song_name,'Artist_name':Artist_name,'Last_week_rank':Last_week_rank,'Peak_rank':Peak_rank,'Weeks_on_board':Weeks_on_board})
data.head() 
```

Question No. 9 Scrape the details most watched tv series of all time from imdb.com.

You have to find the following details:
    
A) Name

B) Year span

C) Genre

D) Run time

E) Ratings

F) Votes

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)
url = "https://www.imdb.com/list/ls095964455"
driver.get(url)

Name=[]
try:
    element=driver.find_elements_by_xpath("//h3[@class='lister-item-header']//a")
    for i in element:
        Name.append(i.text)
except NoSuchElementException as e:
    print("Exception Raised:",e)
    
Year=[]
try:
    element=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
    for i in element:
        Year.append(i.text)
except NoSuchElementException as e:
    print("Exception Raised:",e)
    
Genre=[]
try:
    element=driver.find_elements_by_xpath("//span[@class='genre']")
    for i in element:
        Genre.append(i.text)
except NoSuchElementException as e:
    print("Exception Raised:",e)
    
Run_time=[]
try:
    element=driver.find_elements_by_xpath("//span[@class='runtime']")
    for i in element:
        Run_time.append(i.text)
except NoSuchElementException as e:
    print("Exception Raised:",e)
    
Rating=[]
try:
    element=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']")
    for i in element:
        Rating.append(i.text)
except NoSuchElementException as e:
    print("Exception Raised:",e)
    
Votes=[]
try:
    element=driver.find_elements_by_name("nv")
    for i in element:
        Votes.append(i.text)
except NoSuchElementException as e:
    print("Exception Raised:",e)
    
time.sleep(3)    
data = pd.DataFrame({'Name':Name,'Year':Year,'Genre':Genre,'Run_time':Run_time,'Rating':Rating,'Votes':Votes})
data.head() 
```

Question No. 1 Scrape the details of most viewed videos on YouTube from Wikipedia:
    
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
    
You need to find following details:
    
A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
driver=webdriver.Chrome('chromedriver.exe')
time.sleep(4)
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
time.sleep(5)

Elements=[]
try:
    element=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//td")
    for i in element:
        Elements.append(i.text.split(','))
except NoSuchElementException as e:
    print("Exception Raised:",e)

RANK=[]
RANK=Elements[0::6]

NAME=[]
NAME=Elements[1::6]

views=[]
try:
    element=driver.find_elements_by_xpath("//td[@align='center']")
    for i in element:
        views.append(i.text)
except NoSuchElementException as e:
    print("Exception Raised:",e)

VIEWS=[]
VIEWS=views[1::3]

Uploads_Date=[]
try:
    element=driver.find_elements_by_xpath("//td[@align='right']")
    for i in element:
        Uploads_Date.append(i.text)
except NoSuchElementException as e:
    print("Exception Raised:",e)
    
time.sleep(3)    
data = pd.DataFrame({'RANK':RANK[:10],'NAME':NAME[:10],'VIEWS':VIEWS[:10],'Uploads_Date':Uploads_Date[:10]})
data.head() 
```
