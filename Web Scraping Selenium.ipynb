Q2: Write a python program to scrape data for “Data Scientist” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, full job-description. You have to scrape first 10 jobs data.

```python
search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys('Bangalore')
```

```python
search_btn=driver.find_element_by_xpath("//button[@class='btn']")
search_btn.click()
```

```python
titles=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
job_titles=[]
for i in titles:
    job_titles.append(i.text)
job_titles
```

```python
job_location=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
job_loc=[]
for i in job_location:
    job_loc.append(i.text)
job_loc
```

```python
Company_name=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
company=[]
for i in Company_name:
    company.append(i.text)
company
```

```python
Job_description=[]
url=driver.find_elements_by_xpath("//div[@class='info fleft']//a")
for i in url:
    Job_description.append(i.get_attribute('href'))
Job_description
```

```python
job_desc=[]
for i in Job_description:
    driver.get(i)
    title=driver.find_elements_by_xpath("//section[@class='job-desc']")
    for i in title:
        if i.text is None:
            job_desc.append("No_Job_Description")
        else:
            job_desc.append(i.text)
```

```python
Jobs=pd.DataFrame({})
Jobs['titles']=job_titles[:10]
Jobs['titles_location']=job_loc[:10]
Jobs['Company_name']=company[:10]
Jobs['Job_description']=job_desc[:10]
Jobs.head(10)
```

Q6 : Scrape data of first 100 sunglasses listings on flipkart.com.

```python
urls='https://www.flipkart.com/'
driver.get(urls)
```

```python
search_job=driver.find_element_by_name('q')
search_job.send_keys('Sunglasses')
```

```python
search_btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()
```

```python
page_urls=[]
url=driver.find_elements_by_xpath("//nav[@class='yFHi8N']//a")
for i in url:
    page_urls.append(i.get_attribute('href'))
page_urls
```

```python
Brands_tags=[]
for i in page_urls:
    driver.get(i)
    Brands=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    for i in Brands:
        Brands_tags.append(i.text)
Brands_tags
```

```python
Product_description=[]
for i in page_urls:
    driver.get(i)
    Description=driver.find_elements_by_xpath("//a[@class='IRpwTa']")
    for i in Description:
        Product_description.append(i.text)
Product_description
```

```python
Price_tags=[]
for i in page_urls:
    driver.get(i)
    Price=driver.find_elements_by_xpath("//div[@class='_30jeq3']")
    for i in Price:
        Price_tags.append(i.text)
Price_tags
```

```python
Discount_tags=[]
for i in page_urls:
    driver.get(i)
    Discount=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']//span")
    for i in Discount:
        Discount_tags.append(i.text)
Discount_tags
```

```python
Sunglasses=pd.DataFrame({})
Sunglasses['Brands']=Brands_tags[:100]
Sunglasses['Description']=Product_description[:100]
Sunglasses['Price']=Price_tags[:100]
Sunglasses['Discount']=Discount_tags[:100]
Sunglasses.head()
```
